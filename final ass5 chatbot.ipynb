```python
import nltk
from nltk.chat.util import Chat, reflections

pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, How can I assist you today?"]
    ],
    [
        r"hi|hello|hey there|hola",
        ["Hello! How can I assist you today?"]
    ],
    [
        r"what is your name ?",
        ["I am a COVID-19 bot. How can I assist you today?"]
    ],
    [
        r"how are you ?",
        ["I'm just a bot, but thank you for asking! How can I assist you today?"]
    ],
    [
        r"sorry (.*)",
        ["It's alright.", "No problem.", "Apologies accepted."]
    ],
    [
        r"(.*) precautions(.*)",
        ["To prevent the spread of COVID-19, practice good hygiene such as washing hands frequently, wearing masks, and maintaining social distancing."]
    ],
    [
        r"is vaccine ready ?",
        ["Yes, vaccines are available to combat COVID-19."]
    ],
    [
        r".*vaccine.*",
        ["Vaccine is ready."]
    ],
    [
        r"(.*) cases count(.*)",
        ["The total number of COVID-19 cases is constantly changing. It's best to check the official health department website for the latest updates in your region."]
    ],
    [
        r"(.*) deaths(.*)",
        ["I'm sorry to hear that. COVID-19 has unfortunately caused many deaths worldwide. It's important to follow safety guidelines to prevent further spread."]
    ],
    [
        r"(.*) symptoms(.*)",
        ["Common symptoms of COVID-19 include fever, cough, difficulty breathing, loss of taste or smell, fatigue, and body aches. If you experience these symptoms, it's advisable to seek medical attention."]
    ],
    [
        r"quit",
        ["Thank you for using our COVID-19 bot. Stay safe and take care!"]
    ],
    [
        r"(.*)",
        ["I apologize, but I am a COVID-19 bot and may not have the specific information you're looking for. If you have any COVID-19 related questions, feel free to ask."]
    ]
]

def chat():
    print("Hello! I am a COVID-19 bot. How can I assist you today?")
    chat = Chat(pairs, reflections)
    chat.converse()

if __name__ == "__main__":
    chat()
```
